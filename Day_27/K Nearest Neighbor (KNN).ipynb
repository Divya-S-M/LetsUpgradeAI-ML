```python
import pandas as pd
dataset=pd.read_csv("E:/LetsUpgrade/Day_24/train.csv")
from sklearn import preprocessing
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn import neighbors
```

## Preprocessing

```python
le=preprocessing.LabelEncoder()
le.fit(dataset["Sex"])
dataset["Sex"]=le.transform(dataset["Sex"])
le.fit(dataset["Embarked"])
dataset["Embarked"]=le.transform(dataset["Embarked"])
```

## Splitting Dataset

```python
y=dataset["Pclass"]
X=dataset.drop(["Survived","PassengerId","Name","Ticket","Cabin"],axis=1)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

## Applying KNN 

```python
def funcKNN(k):
    knn=neighbors.KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train).score(X_test,y_test)
    y_pred=knn.predict(X_test)
    print("\nConfusion matrix when k = ",k,":\n",confusion_matrix(y_test,y_pred))
    print("Accuracy is when k = ",k,":",accuracy_score(y_test,y_pred,normalize=True))
    
for i in range(1,268):
    funcKNN(i)
```

## Result

**Inference :**
* The accuracy when k=1 is 90.26% and accuracy when k=267, the accuracy is 71.91%
* The accuracy score of the model decreases as the K value increases.
* The accuracy is maximum when k=1
* Hence, the best fitting k value is 1
